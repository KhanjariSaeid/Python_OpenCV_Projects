```python
#Color slicing using Cube

import numpy as np
import cv2

#Distance function to detemin the distance between points
def distance_in_3D(x1,y1,z1,x2,y2,z2):
    distance=((x1 - x2) ** 2 + (y1 -y2) ** 2+(z1-z2)**2) ** 0.5
    return distance

img=cv2.imread('redstop.jpg')
w,h,z=img.shape

#The side size of the cube
W=0.9*255

#The color to be sliced
(B0,G0,R0)=(0,0,255)
img_red=img.copy()

#Slice the red color and transform other colors to gray
for i in range(w):
    for j in range(h):
        B=img[i,j][0]
        G=img[i,j][1]
        R=img[i,j][2]

        g=0.21*B+0.72*G+0.07*R
        if np.abs(distance_in_3D(B0,G0,R0,B,G,R))>W/2:
            img_red[i,j]=(g,g,g)
        else:
            img_red[i,j]*=1            

cv2.imshow('original',img)
cv2.imshow('red_excluded_by_cube',img_red)
cv2.waitKey(0)
cv2.destroyAllWindows()  
```
